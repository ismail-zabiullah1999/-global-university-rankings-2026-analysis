# üéì **Introduction ‚Äî Global University Rankings 2026 Data Analysis**

This dataset contains detailed information on global university rankings for the year 2026. It includes institutional profiles, regional distribution, and performance indicators such as academic reputation, research output, faculty‚Äìstudent ratio, international diversity, and sustainability.

The objective of this analysis is to perform **data exploration, cleaning, and insight generation** to understand the key factors influencing overall university rankings and identify patterns across regions and institution types.

```python
## üîπ **A. Region-wise Insights**
##**Objective:** Compare performance of universities across regions.


region_group = (uni_data.groupby('Region')[['Overall SCORE', 'AR SCORE', 'Research']].mean().sort_values(by='Overall SCORE', ascending=False))
print("üåç Average Scores by Region:\n", region_group)
print("-" * 60)

```

**Insights:**

* üá™üá∫ **Europe** maintains balanced scores across AR (Academic Reputation) and Overall performance.

* üåç **Africa & South America** might have fewer top-ranking universities, revealing regional disparities.

### üé® **Data Visualization ‚Äî Understanding Performance Patterns**

Visualizations help transform data insights into intuitive, story-driven visuals.
They reveal trends, outliers, and relationships that raw tables can‚Äôt show.

We‚Äôll use Matplotlib and Seaborn for static plots (easy + notebook-friendly).


```python
##üîπ 1Ô∏è‚É£ Bar Chart ‚Äî Top 10 Institutions by Overall SCORE
##üéØ Purpose: Highlight the highest-performing universities globally.


top10 = uni_data.sort_values(by='Overall SCORE', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=top10, x='Overall SCORE', y='Institution Name', palette='Blues_r')
plt.title('üèÜ Top 10 Universities by Overall SCORE')
plt.xlabel('Overall SCORE')
plt.ylabel('Institution Name')
plt.show()
```


### üèÜ 1Ô∏è‚É£ Top 10 Universities by Overall SCORE

**Insight:**

* The **top-performing universities** consistently have **very high Academic (AR)** and **Research (Research)** scores.
* **Institutions from the US and UK** dominate the top 10, highlighting the strength of their research ecosystems and faculty‚Äìstudent ratios.
* The **Overall SCORE gap** between top 3 and the rest is significant, suggesting a clear **performance hierarchy** among global universities.

---


```python

# 2Ô∏è‚É£ Heatmap ‚Äî Correlation Between Score Types

corr = uni_data.select_dtypes('number').corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt='.2f')
plt.title('üî• Correlation Between Ranking Metrics')
plt.tight_layout()
plt.show()

```

### üî• 2Ô∏è‚É£ Correlation Between Score Types (Heatmap)

**Insight:**

* **Strong positive correlation** between `AR SCORE` (Academic Reputation), `ER SCORE` (Employer Reputation), and `Overall SCORE` ‚Üí academic and employer prestige drive rank.
* `Research`, `FSR SCORE`, and `CPF SCORE` also show **moderate-to-high correlation**, emphasizing the importance of research output and faculty ratios.
* `SUS SCORE` (Sustainability) and `EO SCORE` (Employment Outcomes) have **lower correlations**, indicating they‚Äôre **independent differentiators** that don‚Äôt always align with academic prestige.

---


```python
# 3Ô∏è‚É£ Box Plot ‚Äî Overall SCORE Distribution by Region


plt.figure(figsize=(10,6))
sns.boxplot(data=uni_data, x='Region', y='Overall SCORE', palette='Set2')
plt.title('üåç Distribution of Overall SCORE by Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```


### üåç 3Ô∏è‚É£ Score Distribution by Region (Box Plot)

**Insight:**

* **Europe and North America** show **higher median Overall Scores**, reflecting concentrated academic excellence.
* **Asia-Pacific** displays **wider score spread**, meaning the region has both top-tier and mid-range universities ‚Äî growing diversity in educational quality.
* **Emerging regions** (e.g., Africa, Latin America) show **lower overall median scores**, but several **outliers** signal rising institutions.

---

```python

# 4Ô∏è‚É£ Scatter Plot ‚Äî AR SCORE vs ER SCORE (Academic vs Employer Reputation)


plt.figure(figsize=(8,6))
sns.scatterplot(data=uni_data, x='AR SCORE', y='ER SCORE', hue='Region', alpha=0.7)
plt.title('üéì Academic vs Employer Reputation')
plt.xlabel('AR SCORE (Academic Reputation)')
plt.ylabel('ER SCORE (Employer Reputation)')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### üéì 4Ô∏è‚É£ Academic vs Employer Reputation (Scatter Plot)

**Insight:**

* A **strong linear trend** appears ‚Äî universities with high `AR SCORE` often have high `ER SCORE`, reinforcing that **academic reputation boosts employability perception**.
* Some institutions, however, **deviate** ‚Äî indicating universities with strong academic focus but less employer visibility (or vice versa).
* **Region clusters** highlight geographic reputation biases (e.g., US/UK institutions grouped in upper-right quadrant).

---

```python

# 5Ô∏è‚É£ Trend Chart ‚Äî Rank Improvement: Previous Rank ‚Üí 2026 Rank


uni_data['2026 Rank'] = pd.to_numeric(uni_data['2026 Rank'], errors='coerce')
uni_data['Previous Rank'] = pd.to_numeric(uni_data['Previous Rank'], errors='coerce')


uni_data = uni_data.dropna(subset=['2026 Rank', 'Previous Rank'])

trend = uni_data.sort_values(by='2026 Rank').head(20)

plt.figure(figsize=(10,6))
plt.plot(trend['Institution Name'], trend['Previous Rank'], 'o--', label='Previous Rank')
plt.plot(trend['Institution Name'], trend['2026 Rank'], 'o-', label='2026 Rank')
plt.xticks(rotation=90)
plt.xlabel('Institution')
plt.ylabel('Rank (Lower is Better)')
plt.title('Rank Improvement Trend: Previous Rank ‚Üí 2026 Rank')
plt.legend()
plt.tight_layout()
plt.show()

```

### üìà 5Ô∏è‚É£ Rank Improvement Trend (Previous Rank ‚Üí 2026 Rank)

**Insight:**

* Several universities show **notable improvement**, suggesting **strong recent performance** or institutional investment.
* A few show **decline in rank**, potentially due to changes in methodology or global competition.
* The visualization helps identify **fast-rising institutions** worth deeper investigation for performance drivers.
